```python
import pandapower as pp
import pandapower.networks as pn


net = pn.case14()
pp.runpp(net, numba=False)
baseline = net.res_bus.vm_pu.copy()


batt = pp.create_storage(net, bus=8, p_mw=-8, max_e_mwh=32, name="Battery")


pp.runpp(net, numba=False)
v_discharge = net.res_bus.vm_pu[8]


net.storage.at[batt, "p_mw"] = 8
pp.runpp(net, numba=False)
v_charge = net.res_bus.vm_pu[8]

print(f"Bus 8 baseline:          {baseline[8]:.4f}")
print(f"Bus 8 battery DISCHARGE: {v_discharge:.4f}   (change {v_discharge - baseline[8]:+.4f})")
print(f"Bus 8 battery CHARGE:    {v_charge:.4f}   (change {v_charge - baseline[8]:+.4f})")
```

```python
import pandapower as pp
import pandapower.networks as pn

def fresh():
    """Load a clean IEEE 14-bus case and solve the base power flow."""
    net = pn.case14()
    pp.runpp(net, numba=False)
    return net

def report(net, label):
    """Print bus 8, bus 13 voltages and the slack's output for one scenario."""
    slack = net.res_ext_grid.p_mw.sum()
    print(f"{label:<34} bus8={net.res_bus.vm_pu[8]:.4f}  "
          f"bus13={net.res_bus.vm_pu[13]:.4f}  slack={slack:7.2f} MW")


net = fresh()
report(net, "1. baseline")


net = fresh()
pp.create_sgen(net, 13, p_mw=5, name="Solar")
pp.runpp(net, numba=False)
report(net, "2. + solar (5MW @13)")


net = fresh()
pp.create_sgen(net, 13, p_mw=5, name="Solar")
b = pp.create_storage(net, 8, p_mw=-8, max_e_mwh=32, name="Battery")
pp.runpp(net, numba=False)
report(net, "3. + solar + battery DISCHARGE")


net.storage.at[b, "p_mw"] = 8
pp.runpp(net, numba=False)
report(net, "4. + solar + battery CHARGE")
```

```python
import matplotlib.pyplot as plt
import pandapower as pp
import pandapower.networks as pn

def scenario(add=None):
    """Build a fresh 14-bus case, optionally add device(s), solve, return the 14 bus voltages."""
    net = pn.case14()
    pp.runpp(net, numba=False)
    if add:
        add(net)
        pp.runpp(net, numba=False)
    return net.res_bus.vm_pu.values


baseline = scenario()

def add_solar(net):
    pp.create_sgen(net, 13, p_mw=5, name="Solar")

def add_solar_discharge(net):
    pp.create_sgen(net, 13, p_mw=5, name="Solar")
    pp.create_storage(net, 8, p_mw=-8, max_e_mwh=32, name="Battery")   

def add_solar_charge(net):
    pp.create_sgen(net, 13, p_mw=5, name="Solar")
    pp.create_storage(net, 8, p_mw=8, max_e_mwh=32, name="Battery")    

solar     = scenario(add_solar)
discharge = scenario(add_solar_discharge)
charge    = scenario(add_solar_charge)


buses = range(14)
plt.figure(figsize=(10, 6))
plt.plot(buses, baseline,  "o-", label="Baseline")
plt.plot(buses, solar,     "s-", label="+ Solar (bus 13)")
plt.plot(buses, discharge, "^-", label="+ Solar + Battery DISCHARGE (bus 8)")
plt.plot(buses, charge,    "v-", label="+ Solar + Battery CHARGE (bus 8)")


plt.axhline(1.05, color="red", ls="--", lw=1, label="limits (0.95-1.05)")
plt.axhline(0.95, color="red", ls="--", lw=1)

plt.xlabel("Bus number")
plt.ylabel("Voltage (pu)")
plt.title("IEEE 14-Bus Voltage Profile: Solar + Battery Scenarios")
plt.xticks(list(buses))
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("voltage_profile.png", dpi=130)   
plt.show()
```
